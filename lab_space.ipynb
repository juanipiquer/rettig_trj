```python
import pdfplumber


pdf_file_path = "Informe-Rettig-tomo1.pdf"

def extract_text_from_pdf(pdf_file_path, start_page, end_page):
    with pdfplumber.open(pdf_file_path) as pdf:
        text = ""
        for page_num in range(start_page - 1, end_page):  # Adjusting to 0-based index
            page = pdf.pages[page_num]
            text += page.extract_text()
    return text


# Extract text from pages 110 to 210
# solo la rm, para dividir entre "RM" y "el resto"
start_page = 132
end_page = 252
pdf_text = extract_text_from_pdf(pdf_file_path, start_page, end_page)

```

```python

def extract_text_and_formats_from_pdf(pdf_file_path, start_page, end_page):
    with pdfplumber.open(pdf_file_path) as pdf:
        text_plain = ""
        text_with_formatting = ""
        for page_num in range(start_page - 1, end_page):  # Adjusting to 0-based index
            page = pdf.pages[page_num]
            for obj in page.extract_words():
                text_plain += obj["text"] + " "
                if "fontname" in obj and "italic" in obj["fontname"].lower():
                    text_with_formatting += f"*{obj['text']}* "
                elif "fontname" in obj and "bold" in obj["fontname"].lower():
                    text_with_formatting += f"**{obj['text']}** "
                else:
                    text_with_formatting += obj["text"] + " "
            text_plain += "\n"  # Add newline at the end of each page
            text_with_formatting += "\n"  # Add newline at the end of each page
    return text_plain.strip(), text_with_formatting.strip()

# Define the path to the PDF file
pdf_file_path = "Informe-Rettig-tomo1.pdf"

# Extract both plain text and text with formatting from pages 110 to 210
start_page = 132
end_page = 252
plain_text, text_with_formatting = extract_text_and_formats_from_pdf(pdf_file_path, start_page, end_page)

# Print both plain text and text with formatting
print("Plain Text:")
print(plain_text)

print("\nText with Formatting:")
print(text_with_formatting)
```

```python

from typing import Container
from io import BytesIO
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import HTMLConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage

def convert_pdf_to_html(path: str, start_page: int, end_page: int) -> str:
    """Converts a range of pages from a PDF file to HTML format."""
    rsrcmgr = PDFResourceManager()
    retstr = BytesIO()
    laparams = LAParams()
    device = HTMLConverter(rsrcmgr, retstr, laparams=laparams)
    fp = open(path, "rb")
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    for page_num, page in enumerate(PDFPage.get_pages(fp), start=1):
        if page_num < start_page:
            continue
        if page_num > end_page:
            break
        interpreter.process_page(page)
    html = retstr.getvalue().decode()
    fp.close()
    device.close()
    retstr.close()
    return html

# Define the path to the PDF file and the range of pages to extract
pdf_file_path = "Informe-Rettig-tomo1.pdf"
start_page = 139
end_page = 252

# Convert PDF pages 110 to 210 to HTML
html_content = convert_pdf_to_html(pdf_file_path, start_page, end_page)

print("html object created")
```

```python
from bs4 import BeautifulSoup
import re
import pandas as pd

def parse_html_content(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    case_id = 1
    current_case_date = None
    current_case_description = ""
    names = []
    case_data = {"case_date": [], "case_description": []}
    
    for tag in soup.find_all(["span", "div"]):
        if tag.text.strip():
            # Check if it's a date (italics)
            if tag.find("span", style="font-family: TimesNewRoman,Italic"):
                if current_case_date:
                    # Store previous case data
                    case_data["case_date"].append(current_case_date)
                    case_data["case_description"].append(current_case_description.strip())
                    current_case_description = ""
                current_case_date = tag.text.strip()
            # Check if it's a name (bold)
            elif tag.find("span", style="font-family: TimesNewRoman,Bold"):
                name = tag.text.strip()
                # Extract last names in all caps
                last_names = re.findall(r'\b[A-Z ]+\b', name)
                if last_names:
                    name_id = f"{case_id}-{name}"
                    names.append(name_id)
            else:
                # Add plain text to case description
                current_case_description += tag.text.strip()
    
    # Add the last case data
    if current_case_date:
        case_data["case_date"].append(current_case_date)
        case_data["case_description"].append(current_case_description.strip())
    
    # Create DataFrame
    name_case_pairs = [(name_id.split('-')[1], name_id.split('-')[0]) for name_id in names]
    df_names = pd.DataFrame(name_case_pairs, columns=['Name', 'Case ID'])
    df_cases = pd.DataFrame(case_data)
    df = pd.merge(df_names, df_cases, left_on='Case ID', right_index=True)
    
    return df



# Parse HTML content and create DataFrame
df = parse_html_content(html_content)

# Display DataFrame
print(df)
```

```python
from bs4 import BeautifulSoup

def extract_names(html_content):
    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all bold elements
    bold_elements = soup.find_all('span', style=lambda value: value and 'font-family: TimesNewRoman,Bold' in value)

    # Extract names
    names = []
    current_name = ""
    for bold_element in bold_elements:
        text = bold_element.get_text().strip()
        # Check if the text contains more than one capitalized word
        if sum(1 for word in text.split() if word.isupper()) > 1:
            # Append the current name to the list of names
            if current_name:
                names.append(current_name)
            current_name = text
        else:
            # Append the text to the current name
            current_name += " " + text

    # Append the last name to the list of names
    if current_name:
        names.append(current_name)

    return names

names = extract_names(html_content)
```

```python

import pandas as pd
from bs4 import BeautifulSoup

def extract_events_and_names(html_content):
    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # Remove all <br> tags
    for br in soup.find_all('br'):
        br.unwrap()
    
    # Find all italic elements containing the date
    italic_elements = soup.find_all('span', style=lambda value: value and 'font-family: TimesNewRoman,Italic' in value)
    
    # Extract events and associated names based on the italicized dates
    events = []
    names = []
    for italic_element in italic_elements:
        # Find the parent div containing the italic element
        parent_div = italic_element.find_parent('div')
        
        if parent_div:
            # Initialize event text and names for the current event
            event_text = ""
            event_names = []
            
            # Iterate over the following sibling divs and spans
            next_sibling = parent_div.find_next_sibling()
            while next_sibling:
                # Check if the sibling is another italicized element
                if 'Italic' in next_sibling.get('style', ''):
                    break
                
                # Append the text of the sibling element to the event text
                event_text += next_sibling.get_text().strip() + " "
                
                # Find bold elements inside the sibling
                bold_elements = next_sibling.find_all('span', style=lambda value: value and 'font-family: TimesNewRoman,Bold' in value)
                for bold_element in bold_elements:
                    event_names.append(bold_element.get_text().strip())
                
                # Move to the next sibling
                next_sibling = next_sibling.find_next_sibling()
            
            # Append the extracted event text to the list of events
            events.append(event_text.strip())
            
            # Append the extracted names to the list of names
            names.append(", ".join(event_names))
    
    return events, names

# Extract events and associated names from the HTML content
events, names = extract_events_and_names(html_content)

# Create a DataFrame
df = pd.DataFrame({
    "Event": events,
    "Names": names
})

# Display the DataFrame
print(df)
```
